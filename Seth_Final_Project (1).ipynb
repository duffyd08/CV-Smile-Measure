```python
import pandas as pd
from PIL import Image
import requests
from io import BytesIO
import time
import os
import glob
import concurrent.futures

def download_nfl_headshot(player_id, image_type="headshot", folder_name="headshots"):
    url = f"https://static.www.nfl.com/image/{image_type}/f_auto,q_auto/league/{player_id}"
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        image_data = response.content
        image = Image.open(BytesIO(image_data))
        filename = f"{folder_name}/{player_id}_{image_type}.png"
        image.save(filename)
        print(f"Downloaded {image_type} for {player_id}")
        return True
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {image_type} for {player_id}: {e}")
        return False
    except Exception as e:
        print(f"Error processing {image_type} for {player_id}: {e}")
        return False

def download_all_headshots_threaded(active_players, folder_name="headshots", max_workers=10):  # Correct parameter name
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:  # Use max_workers
        futures = [executor.submit(download_nfl_headshot, row['player_id'], row['img_type'], folder_name) for _, row in active_players.iterrows()]

        for future in concurrent.futures.as_completed(futures):
            try:
                future.result()
            except Exception as e:
                print(f"An error occurred: {e}")

# Example usage:
folder_name = "/content/drive/MyDrive/nfl_headshots"  # Or any path within your Drive

download_all_headshots_threaded(active_players, folder_name=folder_name, max_workers=20)
```

```python
import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt

def measure_smile(image_path):
    # Load the face detector and facial landmark predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("/content/drive/MyDrive/seth_final/shape_predictor_68_face_landmarks.dat")

    # Read the image
    image = cv2.imread(image_path)
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)

    # Plot the image in Lab color space
    fig, ax = plt.subplots()
    ax.imshow(lab, zorder=5)
    plt.close(fig)
    # Detect faces in the image
    faces = detector(image_gray)

    if len(faces) == 0:
        return 0

    # We'll work with the first face detected
    face = faces[0]

    # Get facial landmarks
    landmarks = predictor(image_gray, face)

    # Extract mouth landmarks (49 to 68)
    mouth_points = []
    for n in range(49, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        mouth_points.append((x, y))

    mouth_points = np.array(mouth_points)

    return mouth_points
```

```python
import os
import cv2
import pandas as pd

image_dir = '/content/drive/MyDrive/seth_final/nfl_headshots'
save_dir = '/content/drive/MyDrive/seth_final/'

if not os.path.exists(image_dir):
    print(f"Image directory not found: {image_dir}")
else:
    image_files = [os.path.join(image_dir, file) for file in os.listdir(image_dir)]

    # Select images from index 405 to 600
    image_files_subset = image_files[405:601]  # Python slicing is exclusive, so end at 601

def process_images(image_files_subset):
    smile_list = []
    filenames = []

    for file in image_files_subset:
        try:
            mouth_points = measure_smile(file)
            smile_list.append(mouth_points)
        except cv2.error:
            smile_list.append(None)

        filenames.append(os.path.basename(file))

    subset_df = pd.DataFrame({'file': filenames, 'mouth_points': smile_list})

    subset_save_path = os.path.join(save_dir, "smile_data_subset_405_600.csv")
    subset_df.to_csv(subset_save_path, index=False)
    print(f"Subset data saved to {subset_save_path}")

process_images(image_files_subset)
```

```python
test_smile = pd.read_csv('/content/drive/MyDrive/seth_final/test_smile.csv')
```

```python
mouth_area_list = []
for i in range(0, len(test_smile)):
    try:
        # 1. Extract the string and split it based on newline characters
        mouth_points_str = test_smile['mouth_points'][i]
        points_list = mouth_points_str.strip('').split('\n')  # Split by newline

        # 2. Clean each point string and convert to integers
        mouth_points = []
        for point_str in points_list:
            point_str = point_str.strip().replace('[', '').replace(']', '')
            if len(point_str) > 0:  # Skip empty strings
                x, y = map(int, point_str.split())  # Split by space and convert to int
                mouth_points.append([x, y])

        # 3. Convert to a NumPy array
        mouth_points = np.array(mouth_points)

        # 4. Now you can perform the calculations (if there are enough points)
        if len(mouth_points) > 12:
            mouth_area = 0.5 * np.abs(np.dot(mouth_points[12:, 0], np.roll(mouth_points[12:, 1], 1)) -
                                      np.dot(mouth_points[12:, 1], np.roll(mouth_points[12:, 0], 1)))
            mouth_area_list.append(mouth_area)
        else:
            print(f"Skipping row {i} due to insufficient mouth points.")
            mouth_area_list.append(0)

    except (TypeError, ValueError, IndexError) as e:
        print(f"Error processing row {i}: {e}")
        mouth_area_list.append(0)

test_smile['smile_area'] = mouth_area_list
```
